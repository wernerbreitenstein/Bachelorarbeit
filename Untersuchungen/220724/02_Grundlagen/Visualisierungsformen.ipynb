
# Visualisierungsformen

**Beispiele zur Erläuterung häufig verwendeter Diagrammarten**

### Import von Bibliotheken und anderen Erweiterungen

```python
from sqlalchemy import create_engine
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Import von Datensets

```python
penguins = sns.load_dataset("penguins")
tips = sns.load_dataset("tips")
```

### Definitionen zur Darstellung der Visualisierungen

```python
sns.set_theme(style='white', font_scale=1.2, palette='Spectral')
```

### Beispiel Histogramm

```python
# Spezifische Definitionen zur Darstellung der Visualisierung
plt.figure(figsize=(16, 8)) # Größe der Visualisierung (in inch)

# Visualisierung des Beispiels
chart = sns.histplot(data=penguins, x="flipper_length_mm", bins=16)

# weitere Anweisungen zur Darstellung der Visualisierung
chart.grid(axis='y')
chart.set_axisbelow(True)
chart.tick_params(left=False, bottom=False)
sns.despine(left=True)
# plt.savefig('charts/Visualisierungsformen_Histplot.pdf', format='pdf')
plt.show()
```

### Beispiel Balkendiagramm

```python
# Spezifische Definitionen zur Darstellung der Visualisierung
plt.figure(figsize=(8, 8)) # Größe der Visualisierung (in inch)

# Visualisierung des Beispiels
chart = sns.barplot(x="day", y="total_bill", data=tips, ci=None)

# weitere Anweisungen zur Darstellung der Visualisierung
chart.grid(axis='y')
chart.set_axisbelow(True)
chart.tick_params(left=False, bottom=False)
sns.despine(left=True)
# plt.savefig('charts/Visualisierungsformen_Barplot.pdf', format='pdf')
plt.show()
```

### Beispiel Boxplot

```python
# Spezifische Definitionen zur Darstellung der Visualisierung
plt.figure(figsize=(16, 2)) # Größe der Visualisierung (in inch)

# Visualisierung des Beispiels
chart = sns.boxplot(x=tips["total_bill"])

# weitere Anweisungen zur Darstellung der Visualisierung
chart.grid(axis='x')
chart.set_axisbelow(True)
chart.tick_params(left=False, bottom=False)
sns.despine(left=True)
# plt.savefig('charts/Visualisierungsformen_Boxplot.pdf', format='pdf')
plt.show()
```
